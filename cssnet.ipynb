```python
def vecfromcolor(c):
    ret = np.zeros(20)
    j = 0
    for i in range(20):
        h = c[i%len(c)]
        ret[i] = (ord(h.lower()) - ord('a')) / float(ord('z'))
    return ret
```

```python
def vecfromhex(c):
    ret = np.zeros(3)
    ret[0] = int(c[1:3], 16) / float(int('ff', 16)) 
    ret[1] = int(c[3:5], 16) / float(int('ff', 16))
    ret[2] = int(c[5:7], 16) / float(int('ff', 16))
    return ret



def hexfromvec(v):
    return 'rgb('+','.join([str(int(min( max( round(255*c), 0), 255))) for c in v])+')'
```
